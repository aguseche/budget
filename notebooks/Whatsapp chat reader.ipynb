# Import and clean data

```python
#Imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
plt.style.use('bmh') #estetica de graficos
plt.rcParams['font.family'] = "serif"
```

```python
#Import txt that contains all the messages, set headers to None
df = pd.read_csv('./_chat.txt', header=None)
df
```

```python
#Drop unnecessary columns and rows
df.drop(df.columns[[1,2]], axis = 1, inplace = True) 
df = df.iloc[1:]
df
```

```python
#Separate into 5 different columns
df = df[0].str.split(' ', n =5, expand=True)
df
```

```python
df.rename(columns={0:'Date', 1:'Time', 2: 'User', 3:'Price', 4:'Type', 5:'Description'}, inplace=True)
df
```

### use drop now bc i think its better for performance

```python
#Set Price type to float64
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df
```

```python
#Delete rows where user, price or type are none / nan
df = df.mask(df.eq('None')).dropna(subset=['User','Price','Type'])
df
```

### Now i join change the format in columns

```python
#Join Date and Time columns and Drop the remainings
df['DateTime'] = df['Date'] + ' ' + df['Time']
df.drop(['Date','Time'], axis = 1, inplace = True) 
df
```

```python
#rephrase date 
df['DateTime'] = df['DateTime'].str.replace('[', '')
df['DateTime'] = df['DateTime'].str.replace(']', '')
df
```

```python
#Convert type to datetime
df['DateTime'] = pd.to_datetime(df['DateTime'], errors="coerce")
df
```

```python
#df = df.set_index(['DateTime'])
df
```

```python
#Rephrase user
df['User'] = df['User'].str.replace(':','')
df
```

```python
#Set Type and Description to Lower Case
df['Type'] = df['Type'].str.lower()
df['Description'] = df['Description'].str.lower()
df
```

# Visualization 

## Total spent per user

```python
#define variables

distinct_users = df['User'].unique()
y_pos = np.arange(len(distinct_users))
totalspent_peruser = df.groupby(['User'])['Price'].sum()
users_colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
```

```python
#Plot barchart
plt.bar(y_pos, totalspent_peruser, color = users_colors)
plt.xticks(y_pos, distinct_users)
plt.ylabel('$ Pesos Argentinos')
plt.title('Amount of money spent by person')
plt.show()
```

```python
#Donut pie chart
fig1, ax1 = plt.subplots()
ax1.pie(totalspent_peruser, colors=users_colors, labels=distinct_users, autopct='%1.1f%%', 
        startangle=90, pctdistance=0.85, radius=1)

#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

## Total spent per type 

```python
#Define variables
types_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
```

```python
df.groupby('Type').sum()['Price'].plot.bar(color = types_colors)
plt.xticks(rotation=0)
plt.ylabel('$ Pesos Argentinos')
plt.xlabel('')
plt.title('Amount of money spent per type')

plt.show()
```

```python
distinct_types = df['Type'].unique()
#Needed bc the groupby.sum orders the unique types alphabetically
distinct_types.sort()
```

```python
#df.groupby('Type').sum()['Price'].plot(kind = "pie", colors = types_colors)
plt.pie(df.groupby('Type').sum()['Price'], labels=distinct_types, autopct='%1.1f%%', colors = types_colors,startangle=90, pctdistance=0.85)

#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
